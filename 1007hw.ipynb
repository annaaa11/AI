```python
import torch
def f2(x, y):
  # трьохгорбна функція
  return 2*x**2  - 1.05*x**4 + 1/6*x**6 + x*y  + y**2

result_true = 0
```

```python
from torch.optim import SGD

# підкоючаємо процесор
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

# вибираємо початкове значення
x = torch.tensor(1.1)
x = x.to(device)
x.requires_grad = True

y = torch.tensor(1.1)
y = y.to(device)
y.requires_grad = True

# швидкість навчання
learning_rate = 1e-2  # 10^-3 або 0.001

# створення алгоритму
optimizer = SGD(
    [x, y],   # список параметрів(тензори з градієнтом)
    lr = learning_rate
)

for t in range(5000):
    # обчислюємо функцію
    result = f2(x, y)
    
    # обчислюємо наскільки y відрізняється від 3
    error = (result_true - result)**2
    
    # обчислюємо всі потрібні градієнти
    error.backward()  # в x з'являється градієнт який робить error меншим
    
    # градієнтний спуск
    optimizer.step()
    optimizer.zero_grad()  # очистити градієнти
    
    # вивід значення функції
    if t % 500 == 1: # кожен сотий крок
        print(f"{error = }")
```
